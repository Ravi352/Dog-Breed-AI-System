```python
df = df.rename(columns={'Unnamed: 0': 'breed'})
df = df.reset_index(drop=True)
```

```python
df.head()
```

```python
from langchain.tools import tool,Tool
from sentence_transformers import SentenceTransformer, util

# Load Sentence Transformer model
model = SentenceTransformer("all-MiniLM-L6-v2")

# Encode breed descriptions from dataset
breed_descriptions = df["description"].tolist()
description_embeddings = model.encode(breed_descriptions, convert_to_tensor=True)

@tool
def get_breed_description(query):
    """Finds the most relevant dog breed based on semantic similarity."""
    query_embedding = model.encode(query, convert_to_tensor=True)

    # Compute cosine similarity
    similarities = util.pytorch_cos_sim(query_embedding, description_embeddings)
    best_match_idx = similarities.argmax().item()

    return df.iloc[best_match_idx]["description"]

# Define LangChain tool
semantic_search_tool = Tool(
    name="Breed Description Search",
    func=get_breed_description,
    description="Provides breed descriptions based on user queries."
)
```

```python
semantic_search_tool.invoke("Which Dog is good for Harsh weather ?")
```

```python
from langchain_experimental.agents import create_pandas_dataframe_agent
from langchain_community.llms import Ollama  
from langchain_ollama.llms import OllamaLLM
import pandas as pd

llm = OllamaLLM(model="llama3.2:1b")


pandas_agent = create_pandas_dataframe_agent(
    llm, df,allow_dangerous_code=True,handle_parsing_errors=True
)
```

```python
pandas_agent.invoke('I want a dog which is easy to handle and is very cute. Which dog you would refer and tell me the grooming value also ')
```
