```python
df.groupby('group')['grooming_frequency_category'].apply(lambda x: x.isna().sum())
```

```python
df.groupby('grooming_frequency_category')['group'].count()
```

```python
## Filling the nan value in the grooming_frequency_category we will group the data on the basis of group and replace nan value with the most occuring
## grooming_frequency_category for the group

df['grooming_frequency_category'] = df.groupby('group')['grooming_frequency_category'].transform(lambda x:x.fillna(x.mode()[0]))
```

```python
df['grooming_frequency_category'].isnull().sum()
```

```python
df.columns
```

```python
## Similar concept I will be using for the other categorical data 
df['shedding_category'] = df.groupby('group')['shedding_category'].transform(lambda x:x.fillna(x.mode()[0]))
df['energy_level_category'] = df.groupby('group')['energy_level_category'].transform(lambda x:x.fillna(x.mode()[0]))
df['trainability_category'] = df.groupby('group')['trainability_category'].transform(lambda x:x.fillna(x.mode()[0]))
df['demeanor_category'] = df.groupby('group')['demeanor_category'].transform(lambda x:x.fillna(x.mode()[0]))
```

```python
df.isnull().sum()
```

```python
df['grooming_frequency_value'] = df.groupby('grooming_frequency_category')['grooming_frequency_value'].transform(lambda x:x.fillna(x.mean()))
```

```python
df['grooming_frequency_value'] .isnull().sum()
```

```python
df['shedding_value'] = df.groupby('shedding_category')['shedding_value'].transform(lambda x:x.fillna(x.mean()))
df['energy_level_value'] = df.groupby('energy_level_category')['energy_level_value'].transform(lambda x:x.fillna(x.mean()))
df['trainability_value'] = df.groupby('trainability_category')['trainability_value'].transform(lambda x:x.fillna(x.mean()))
df['demeanor_value'] = df.groupby('demeanor_category')['demeanor_value'].transform(lambda x:x.fillna(x.mean()))
```
